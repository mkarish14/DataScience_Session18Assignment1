```python
#Problem 2

import statistics as st

Group1 = [51, 45, 33, 45, 67]
Group2 = [23, 43, 23, 43, 45]
Group3 = [56, 76, 74, 87, 56]

allgroups= Group1 + Group2 + Group3

sum_Group1 = sum(Group1)
sum_Group2 = sum(Group2)
sum_Group3 = sum(Group3)

mean_Group1 = st.mean(Group1)
mean_Group2 = st.mean(Group2)
mean_Group3 = st.mean(Group3)

print("Sum of Group 1:",sum_Group1)
print("Mean of Group 1:",mean_Group1)

print("Sum of Group 2:",sum_Group2)
print("Mean of Group 2:",mean_Group2)

print("Sum of Group 3:",sum_Group3)
print("Mean of Group 3:",mean_Group2)

Totalsum=  sum(allgroups)
Totalmean= st.mean(allgroups)

print("Total Sum: ",Totalsum)
print("Total Mean: ",Totalmean)



#Calculating sum of squares
d_Group1 = []
d_Group2 = []
d_Group3 = []

d2_Group1 = []
d2_Group2 = []
d2_Group3 = []

d_allgroups =[]
d2_allgroups =[]

for i in Group1:
    diff =0
    diff=i-mean_Group1
    d_Group1.append(diff)

for i in d_Group1:
    d2 = i*i
    d2_Group1.append(d2)

    
for i in Group2:
    diff =0
    diff=i-mean_Group2
    d_Group2.append(diff)

for i in d_Group2:
    d2 = i*i
    d2_Group2.append(d2)

for i in Group3:
    diff =0
    diff=i-mean_Group3
    d_Group3.append(diff)

for i in d_Group3:
    d2 = i*i
    d2_Group3.append(d2)
    
for i in allgroups:
    diff =0
    diff=i-Totalmean
    d_allgroups.append(diff)

for i in d_allgroups:
    d2 = i*i
    d2_allgroups.append(d2)
    
#print(d_Group1)
#print(d2_Group1)

#print(d_Group2)
#print(d2_Group2)

#print(d_Group3)
#print(d2_Group3)
    
sum_sq1 = sum(d2_Group1)
mean_sq1 =st.mean(d2_Group1)

sum_sq2 = sum(d2_Group2)
mean_sq2 =st.mean(d2_Group2)

sum_sq3 = sum(d2_Group3)
mean_sq3 =st.mean(d2_Group3)

sum_sqallgroups = sum(d2_allgroups)
mean_sqallgroups = st.mean(d2_allgroups)

#print(sum_sq1)
#print(mean_sq1)


#print(sum_sq2)
#print(mean_sq2)

#print(sum_sq3)
#print(mean_sq3)

#print(sum_sq3)
#print(mean_sq3)

#print(sum_sqallgroups)
#print(mean_sqallgroups)

#sst =
# ss between
ssb = (((mean_Group1 - Totalmean)*(mean_Group1 - Totalmean)) + ((mean_Group2 - Totalmean)*(mean_Group2 - Totalmean)) + ((mean_Group3 - Totalmean)*(mean_Group3 - Totalmean)))*5
ssw = sum_sq1+sum_sq2+sum_sq3
sst= ssb+ssw
#print(ssw)
#print(ssb)
#print(sst)

#Degrees of freedom
#Total number of observations - 1 --n-1
dft=15-1
#degree of freedom within n-k k-no of groups
dfw= 15- 3

#degree of freedom between  k-1 (no of groups -1)
dfb= 3-1

#calculate msbbetween and msw
msb = ssb/dfb
msw = ssw/dfw

f = msb/msw
#print(msb)
#print(msw)
print("F statistics:",f)

#from the table F=3.89, and our F=9.75 CALC VALUE> TABLE VALUE, therefore we reject null hypothesis and conclude at least 2 means are significatly difference
```
